# Data Wrangling
Data wrangling, also known as data munging, is the process of transforming and mapping raw data into a more useful format for analysis. It involves cleaning, restructuring, and formating (e.g renaming columns and fixing data types) to the data to make it suitable for analysis or visualization. 

## Data Cleaning
in this section we will take a look at creating, renaming and droping columns, type convertion and sorting. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
taxis=pd.read_csv(r"C:\Users\MUNYUA CHARLES\Downloads\2019_Yellow_Taxi_Trip_Data.csv")
```

```python
df=taxis.copy()
```

```python
df.head()
```

```python
#Droping unwanted columns 
mask=df.columns.str.contains('id$|store_and_fwd_flag',regex=True)
columns_to_drop=df.columns[mask]
columns_to_drop
```

```python
df=df.drop(columns=columns_to_drop)
df.head()
```

```python
#Renaming Columns
df=df.rename(
    columns={
        'tpep_pickup_datetime':'pickup',
        'tpep_dropoff_datetime':'dropoff'
    }
)
df.columns
```

```python
#Type Convertion
df.dtypes
```

both pikup and drop off should be of datetime data types

```python
#converting pickup and dropoff from object to date time data types 
df[['pickup','dropoff']]=df[['pickup','dropoff']].apply(pd.to_datetime)
```

```python
df.dtypes
```

there is another way to perform type convertion . for numeric values we use the pd.to_numeric() function, and will see astype() method,which is a more generic method.

### Creating new columns 
we will calculate the following columns 
1. Elapsed time of the trip

2. The trip percentage

3. The total taxes,tolls,fees and surcharges

4. The average speed of the taxi

```python
df['elapsed_time']=df['dropoff']-df['pickup']
df['cost_before_trip']=df['total_amount']-df['tip_amount']
df['trip_pct']=df['tip_amount']/df['cost_before_trip']*100
df['fees']=df['cost_before_trip']-df['fare_amount']
```

```python
df.head(2)
```

```python
# sorting by values 
# we can use the sort_values() method to sort based on any number of columns 
df.sort_values(['passenger_count','pickup'],ascending =[False,True]).head(2)
```

tp pick up the largest and the smallest rows we use nlargest() and nsmallest() instead.Looking at the 3 trips with the longest elapse time

```python
df.nlargest(3,'elapsed_time')
```
