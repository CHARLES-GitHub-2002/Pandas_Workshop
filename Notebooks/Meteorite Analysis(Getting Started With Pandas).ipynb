- Filtering with boolean masks
A boolean mask is a array like structure of boolean values it's a way to specify which rows/columns we want to select (True) and which we don't (False).



This is a boolean mask for metrorites weighing more than 50 grams that were found on earth (i.e they were not observed falling)

```python
mask=(df['mass (g)']>50) & (df.fall == 'Found')
mask
```

**important**: Take note of syntax here.we surround each condition with parentheses, and we use bit wise operators (&,|,~) instead od logic operators ( and , or ,not)
we can use a boolean mask to select the subset of metrorites weiging more than 1 Milion grams

```python
df[(df['mass (g)']>1e6) & (df.fall == 'Fell')]
```

**tip :Boolean masks can be used with loc[] and iloc[]**


```python
#An altenative to this is the query() method
df.query("`mass (g)` > 1e6 and fall == 'Fell'")
```
